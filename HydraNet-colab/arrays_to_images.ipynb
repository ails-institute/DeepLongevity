```python
import numpy as np
import os
from skimage.io import imsave, imshow, imread
import gc
```

```python
def arry2imgs(a, folder):
  a = np.squeeze(a)
  a = np.uint8(np.rint(a*65535))
  os.mkdir(folder)
  for img_name in range(0,a.shape[0]):
    imsave(os.path.join(folder,f"{img_name}.png"),
           np.squeeze(a[img_name, :, :]))
```

```python
main_folder = '/content/drive/My Drive/worm_models/movement/unseg_imgs'
os.mkdir(main_folder)
```

```python
#@title Get dataset unseg
data_file = '/content/drive/My Drive/worm_models/movement/move_data_unseg.npz'
unseg = np.load(data_file, allow_pickle=True)

folder = os.path.join(main_folder, 'train')
os.mkdir(folder)
arry2imgs(np.float16(unseg['imgs_train_high']), os.path.join(folder, 'high'))
arry2imgs(np.float16(unseg['imgs_train_low']), os.path.join(folder, 'low'))

folder = os.path.join(main_folder, 'validate')
os.mkdir(folder)
arry2imgs(np.float16(unseg['imgs_validate_high']), os.path.join(folder, 'high'))
arry2imgs(np.float16(unseg['imgs_validate_low']), os.path.join(folder, 'low'))

folder = os.path.join(main_folder, 'test')
os.mkdir(folder)
arry2imgs(np.float16(unseg['imgs_test_high']), os.path.join(folder, 'high'))
arry2imgs(np.float16(unseg['imgs_test_low']), os.path.join(folder, 'low'))


del unseg

gc.collect()
```
